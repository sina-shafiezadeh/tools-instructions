```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = "D:\\data.xlsx"
img_path = "D:\\images"
```

# Conegliano dataset

```python
df = pd.read_excel(file_path, sheet_name='conegliano dataset')
df.head()
```

```python
sns.set_context("notebook", rc={'axes.labelsize': 20.0, 'xtick.labelsize': 20.0, 'axes.linewidth': 1.25,
                                'ytick.labelsize': 20.0, 'legend.fontsize': 20.0})
#sns.plotting_context() #show possible parameters

custom_params = {"axes.spines.right": False, "axes.spines.top": False,
                 'axes.grid': True, 'grid.linestyle': '--', 'grid.color': '0.8'}

sns.set_theme(context='paper', style='ticks', rc={**sns.plotting_context(), **custom_params})

plt.figure(figsize=(12, 8), dpi=300)


custom_palette = {'RCV': 'darkgray', 'LOO': 'cornflowerblue', 'Cal1': 'yellowgreen', 'Cal2': 'orange'}
violin_plot = sns.violinplot(data=df, x="metric", y="value", hue="model", inner="box", palette=custom_palette,
                             saturation=1, linewidth=0.3, cut=1.5,
                             inner_kws=dict(box_width=8, whis_width=2, color="dimgray",
                             marker='_', markersize=8, markeredgewidth=3, markeredgecolor="blue"))


plt.ylim(0, 115) # Set the y-axis limits
violin_plot.set(xlabel="", ylabel="Percentage (%)")


#move the legend outside the figure
#violin_plot.legend(loc='upper left', bbox_to_anchor=(1, 1)) 
violin_plot.legend(loc='upper center', bbox_to_anchor=(0.5, -0.07), frameon=False,
                   ncol=len(df['model'].unique()), columnspacing=0.5, handlelength=1.0)

plt.savefig(f'{img_path}//results_conegliano_dataset.JPG', bbox_inches='tight', dpi=300)
```
